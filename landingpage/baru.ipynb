```python
import matplotlib.pyplot as plt

# Data sederhana
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Membuat grafik
plt.plot(x, y)
plt.title("Contoh Grafik")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Menampilkan grafik di notebook
plt.show()
```

```python
# script.py
import pandas as pd  # Impor pandas
import matplotlib.pyplot as plt  # Impor matplotlib untuk visualisasi

# Membaca file Excel
file_path = "C:\\Users\\ADVAN\\Downloads\\data_siswa.xlsx"
data = pd.read_excel(file_path)  # Membaca file Excel

# Menampilkan data (opsional)
print(data.head())  # Menampilkan 5 baris pertama

# Menghitung jumlah siswa per sekolah
jumlah_siswa_per_sekolah = data['Asal Sekolah'].value_counts()

# Membuat grafik batang horizontal untuk memperbaiki tampilan label
plt.barh(jumlah_siswa_per_sekolah.index, jumlah_siswa_per_sekolah.values, color='b')
plt.title("Jumlah Siswa Berdasarkan Asal Sekolah")
plt.xlabel("Jumlah Siswa")
plt.ylabel("Asal Sekolah")
plt.tight_layout()
plt.show()
```

```python
# script.py
import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:\\Users\\ADVAN\\Downloads\\data_siswa.xlsx"
data = pd.read_excel(file_path)
jumlah_siswa_per_sekolah = data['Asal Sekolah'].value_counts()
plt.barh(jumlah_siswa_per_sekolah.index, jumlah_siswa_per_sekolah.values, color='b')
plt.title("Jumlah Siswa Berdasarkan Asal Sekolah")
plt.xlabel("Jumlah Siswa")
plt.ylabel("Asal Sekolah")
plt.tight_layout()
plt.savefig("jumlah_siswa.png")
```
